```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.svm import SVC
import seaborn as sns
```

```python
raw_data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/epi_r.csv')
```

```python
df = raw_data.drop(['protein', 'fat', 'sodium', 'calories'],axis = 1)
```

```python
pd.set_option("display.max_rows",999)

df.isnull().sum()
```

```python
a = df.columns.tolist()
for x in a:
    print(x, np.unique(df[x]))
```

```python
df['ratebin'] = np.where(df['rating']<3, 0,1)
```

```python
import us
us.states
States = ["Alabama","Alaska","Arizona","Arkansas","California","Colorado",
  "Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois",
  "Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland",
  "Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana",
  "Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York",
  "North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania",
  "Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah",
  "Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
#states.str(replace('\"', '\''))
States = [States[x].replace('\"', '\'') for x in range(len(States))]
states = [States[x].lower() for x in range(len(States))]
states
```

```python

dfcols= pd.Series(df.columns).tolist()
dfcols
```

```python
# statesel = [for x in states if x in dfcols]
statesel = []
for x in states:
    if x in dfcols:
        statesel.append(x)
dfdrop = df.drop(statesel, axis =1).drop('title',axis = 1)
```

```python
a = dfdrop.columns.tolist()
for x in a:
    print(x, np.unique(dfdrop[x]))
dfdrop.info()
```

```python
from sklearn.svm import SVR
svr = SVR()
X = dfdrop.drop('ratebin', 1).sample(frac=0.3, replace=True, random_state=1)
Y = dfdrop.ratebin.sample(frac=0.3, replace=True, random_state=1)
svr.fit(X,Y)
```

```python
plt.scatter(Y, svr.predict(X))
```

```python
svr.score(X, Y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svr, X, Y, cv=5)
```
